```python
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_median_age'
```

```python
df = pd.read_html(url)
```

```python
# Type of data is a list if picked up using pd.read_html method
type(df)
```

```python
# Each table is stored in the list
# Length of list shows how many tables are stored
len(df)
```

```python
# We will use the second table
df = df[1]
```

```python
df
```

```python
# Checking values in Type column
df.Type.value_counts()
```

```python
# Checking what is SDG Region?

df[df.Type=='SDG region']
```

```python
# What is SDG Region?

df[df.Type=='SDG subregion']
```

```python
# Ok so these are regions of the world, which is good data
# I am not interested in data groups
# Removing rows for which the type is not Country/Area or SDG subregion or SDG region
```

```python
# Creating boolean series for 'Country/Area'
c = df.Type == 'Country/Area'
c
```

```python
# Creating boolean series for 'SDG subregion'
sr = df.Type == 'SDG subregion'
sr
```

```python
# Creating boolean series for 'SDG region'
r = df.Type == 'SDG region'
r
```

```python
# Extracting required data
data = df[c | sr | r] # This means c or sr or r
data
```

```python
# Checking if the unnecessary groups have been eliminated
data.Type.unique()
```

```python
# Hence unwanted rows are removed
data.head()
```

```python
# Simplifying a long column name

data = data.rename(columns={'Region, subregion, country or area':'Name'})
```

```python
data.head()
```

```python
data.Type.unique()
```

```python
# Removing letters SDG
data.loc[:,'Type'] = data.Type.str.replace('SDG region', 'Region')
```

```python
data.loc[:,'Type'] = data.Type.str.replace('SDG subregion', 'Subregion')
```

```python
# Checking if SDG has been removed
data.Type.unique()
```

```python
data.head()
```

```python
# Checking data types
data.dtypes
```

```python
# Convert all year columns to float data type

for i in data.columns:
    if i not in ['Name', 'Type']:
        data[i] = round(data[i].astype(float),2)
```

```python
data.dtypes
```

```python
data.head()
```

```python
# Check for missing values

data.isna().sum()
```

```python
data.head()
```

```python
# Check for duplicates in Name column
data.shape[0]
```

```python
len(data.Name.unique().tolist())
```

```python
# Hence no duplicates exist
# Separating Country/Area data in separate dataframe and regions into separate dataframe

countries = data[data.Type=='Country/Area']
countries.head()
```

```python
# Any data which is not 'Country/Area' to be a separate dataframe

regions = data[data.Type!='Country/Area']
regions
```

```python
# Now reset the index on both dataframes

countries.reset_index(drop=True, inplace=True)
regions.reset_index(drop=True, inplace=True)
```

```python
### Question ###
### How much was the growth of median age in all African countries from 1950-2020

# Strategy to solve this question
# 1. Acquire list of all African countrires from Wikipedia
# 2. Filter those countries' data from original dataframe

data.head()
```

```python
# Scraping list of African countries

african_url = 'https://en.wikipedia.org/wiki/List_of_African_countries_by_population'
africa = pd.read_html(african_url)
```

```python
africa[0].head()
```

```python
africa[0].Country.unique()
```

```python
# Extract african countries list from this dataframe
af_list = africa[0].Country.unique().tolist()
af_list
```

```python
# Changing country names to title case and saving in another list
# This is to ensure name comparison between this list and dataframe countries happens in same case
# Otherwise data quality would compromise and some data might get missed

country_names = []
for i in af_list:
    country_names.append(i.title())
country_names
```

```python
# Now extracting age data using african country list
# Below compares country names in title case and supplies boolean series to dataframe to extract required data

african_age = data[data.Name.str.title().isin(country_names)]
```

```python
african_age
```

```python
# Creating growth dataframe for years 1950 and 2020 only

africa_age_growth = african_age.loc[:,['Name','1950','2020']].set_index('Name')
africa_age_growth
```

```python
# Now adding a column which shows the growth of median age

africa_age_growth['Growth'] = africa_age_growth['2020'] - africa_age_growth['1950']
```

```python
africa_age_growth.head()
```

```python
# Removing 1950 and 2020 columns

africa_age_growth.drop(['1950','2020'],axis=1,inplace=True)
```

```python
africa_age_growth.head()
```

```python
# Plotting data

africa_age_growth.sort_values(by='Growth').plot(kind='bar',\
                                                figsize=(15,10),\
                                                color='red');
```

```python
# Above graph shows that about half of african countries' have decreased their median age
# Togo and Niger had no growth or drop in median age at all
# Sao Tome and Principe had a drop in median age beyond 5 years
# Mauritiaus had biggest increase in median age of all the African countries
```
